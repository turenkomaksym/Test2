```python
import pandas as pd
import re
import warnings
warnings.filterwarnings('ignore')
```

# Let's prepare our first DF

```python
df = pd.read_csv('refused_order_products.csv')
df
```

```python
df10 = pd.DataFrame(df.loc[[0]], columns = df.columns).set_index('lvl')
df13 = pd.DataFrame(df.loc[[3]], columns = df.loc[2,]).set_index('lvl')
df16 = pd.DataFrame(df.loc[[6]], columns=df.loc[5,]).set_index('lvl')
df19 = pd.DataFrame(df.loc[[9]], columns=df.loc[8,]).set_index('lvl')
df24 = pd.DataFrame(df.loc[[12]], columns=df.loc[11,]).set_index('lvl')
df29 = pd.DataFrame(df.loc[[15]], columns=df.loc[14,]).set_index('lvl')
```

```python
def prepare_df(df):
    df = df.loc[:, df.filter(regex='lvl|Refused').columns]
    df.set_axis(df.columns.str.extract(r'\[(\w+)', expand = False), axis = 1, inplace = True)
    df = df.T
    df.index.name = 'id_product'
    df.reset_index(inplace = True)
    return df

df10 = prepare_df(df10)
df13 = prepare_df(df13)
df16 = prepare_df(df16)
df19 = prepare_df(df19)
df24 = prepare_df(df24)
df29 = prepare_df(df29)
```

```python
# Concatenate our dataframes

from functools import reduce

data_frames = [df10,df13,df16,df19,df24,df29]

df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['id_product'],how='outer'), data_frames)
df_merged.head()
```

# Let's prepare our second df

```python
df2 = pd.read_csv('product_info.csv', encoding='windows-1251', sep = '\t', header=1)
df2.head(20)
```

```python
# Let's delete rows 14-17

df2.drop(df2.index[14:18], inplace = True)
df2.reset_index(drop=True, inplace = True)
df2.head(20)
```

# Concatenate two main df's

```python
df_full = pd.merge(df_merged, df2, on = 'id_product')
df_full.drop('Название', axis = 1, inplace = True)
df_full.head()
```

# Рассмотрим динамику изменения по уровням


```python
df_merg2 = df_merged.set_index('id_product')
df_merget_T = df_merg2.T
df_merget_T
```

```python
# difference in values for two adjacement lvl

df_merget_T.fillna('0', inplace = True)
df_merget_T = df_merget_T.astype(int)
df_div = df_merget_T.diff()
# положительное число - на сколько увеличился показатель в новом диапазоне уровней в сравнении с предыдущим диапазоном уровней
# отрицательное число - на сколько уменьшился показатель в новом диапазоне уровней в сравнении с предыдущим диапазоном уровней
```

```python
import plotly.plotly as py
import plotly.graph_objs as go


def plotly_df(df_set):
    df_set = df_set.T
    df_set = df_set.reset_index()
    
    trace = go.Table(
        header=dict(values=list(df_set.columns),
                    fill = dict(color='#C2D4FF'),
                    align = ['center']),
        cells=dict(values=[ df_set[i].values for i in df_set.columns ],
                   fill = dict(color='#F5F8FF'),
                   align = ['center']))
    data = [trace] 
    return py.iplot(data, filename = 'pandas_table')
```

```python
plotly_df(df_div)
```

```python
df_perc = df_merget_T.pct_change().mul(100).round(2)
pd.options.display.float_format = '{:,.2f}'.format
plotly_df(df_perc)

#динамика в процентном отношении
```
