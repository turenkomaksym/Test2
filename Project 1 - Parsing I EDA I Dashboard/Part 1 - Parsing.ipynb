# ЭТО ПОЧТИ РАБОТАЕТ - надо улучшить - PAБОТАЕТ 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


df = pd.DataFrame(columns = ['Title', 'Image', 'Old price', 'New price','Diagonal of display', 'Shape of display', 'Processor series',
                            'CPU', 'Cores', 'CPU base frequency', 'RAM', 'RAM type', 'RAM frequency', 'HDD', 'SSD',
                            'Video card', 'Model of video card', 'Video card capacity', 'Web-camera', 'Shape', 'Weight'])

def get_html(url):
    response = requests.get(url)
    return response.text

def get_total_pages(html):
    soup = BeautifulSoup(html, 'lxml')
    total_pages = soup.find('ul', class_='pagination-number-list').find_all('li')[-1].text
    return int(total_pages)

def get_all_links_from_page(html):
    main_url = 'https://www.foxtrot.com.ua'
    soup = BeautifulSoup(html, 'lxml')
    links = soup.find_all('div', class_='listing-item product-item')
    links = [ main_url+link.find('a').get('href') for link in links ]
    return links

def get_data(link):
    print(link)
    response = requests.get(link).text
    soup = BeautifulSoup(response, 'lxml')
    table = soup.find('ul', class_='points-list').find_all('li')
    data = []
    for tab in table:
        data.append({
            'title':tab.find('p', class_='point__left-text').text.strip(),
            'value':tab.find('p', class_='point__right-text').text.strip()
        })
    
    df.loc[link,'Title'] = soup.find('div', class_='mark').find('h1').text.strip()
    try:
        df.loc[link,'Old price'] = soup.find('div', class_='col-left').find('div', class_='price__not-relevant').find_all('span')[0].text.strip()
    except:
        df.loc[link,'Old price'] = ''
    df.loc[link,'New price'] = soup.find('div', class_='col-left').find('div', class_='price__relevant').find_all('span')[0].text.strip()
    df.loc[link,'Image'] = soup.find('div', class_='img-wrapper').find('img').get('src')
    
    for i in range(len(data)):
        if 'Диагональ дисплея' in data[i]['title']: df.loc[link,'Diagonal of display'] = data[i]['value']
        elif 'Разрешение дисплея' in data[i]['title']: df.loc[link,'Shape of display'] = data[i]['value']
        elif 'Серия процессора' in data[i]['title']: df.loc[link,'Processor series'] = data[i]['value']
        elif 'Модель процессора' in data[i]['title']: df.loc[link,'CPU'] = data[i]['value']
        elif 'Количество ядер' in data[i]['title']: df.loc[link,'Cores'] = data[i]['value']
        elif 'Базовая частота процессора' in data[i]['title']: df.loc[link,'CPU base frequency'] = data[i]['value']
        elif 'Объем ОЗУ' in data[i]['title']: df.loc[link,'RAM'] = data[i]['value']
        elif 'Тип памяти' in data[i]['title']: df.loc[link,'RAM type'] = data[i]['value']
        elif 'Частота памяти' in data[i]['title']: df.loc[link,'RAM frequency'] = data[i]['value']
        elif 'Объем жесткого диска HDD' in data[i]['title']: df.loc[link,'HDD'] = data[i]['value']
        elif 'Объем SSD диска' in data[i]['title']: df.loc[link,'SSD'] = data[i]['value']
        elif 'Производитель видеокарты' in data[i]['title']: df.loc[link,'Video card'] = data[i]['value']
        elif 'Модель видеокарты' in data[i]['title']: df.loc[link,'Model of video card'] = data[i]['value']
        elif 'Объем видеопамяти' in data[i]['title']: df.loc[link,'Video card capacity'] = data[i]['value']
        elif 'Веб-камера' in data[i]['title']: df.loc[link,'Web-camera'] = data[i]['value']
        elif 'Габариты' in data[i]['title']: df.loc[link,'Shape'] = data[i]['value']
        elif 'Вес' in data[i]['title']: df.loc[link,'Weight'] = data[i]['value']


def main():
    url_site = 'https://www.foxtrot.com.ua/ru/shop/noutbuki.html?page=1'
    base_url = 'https://www.foxtrot.com.ua/ru/shop/noutbuki.html?page='
    total_pages = get_total_pages(get_html(url_site))
    print('There are %d pages.' %total_pages)
    
    for page in range(1, total_pages+1):
        print("Parsing %d%% " %(page/total_pages*100))
        url = base_url+str(page)
        links = get_all_links_from_page(get_html(url))
        #print(links)
        for link in links:
            get_data(link)


if __name__ == '__main__':
    main()
```
