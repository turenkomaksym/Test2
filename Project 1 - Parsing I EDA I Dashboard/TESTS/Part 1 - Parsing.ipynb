```python
import pandas as pd


urls = ['https://www.foxtrot.com.ua/ru/shop/noutbuki_lenovo_legion-y520-15-80wk00uxra.html', 
        'https://www.foxtrot.com.ua/ru/shop/noutbuki_hp_pavilion-gaming-15-cx0042ur-4ps16ea.html', 
        'https://www.foxtrot.com.ua/ru/shop/noutbuki_hp_250-g6-3vk20es.html', 
        'https://www.foxtrot.com.ua/ru/shop/noutbuki_apple_a1466-macbook-air-mqd32ua-a.html', 
        'https://www.foxtrot.com.ua/ru/shop/noutbuki_acer_aspire-5-a515-51g-38eg-nx-gvmeu-015.html']

df = pd.DataFrame(columns = ['Title', 'Image', 'Old price', 'New price','Diagonal of display', 'Shape of display', 'Processor series',
                            'CPU', 'Cores', 'CPU base frequency', 'RAM', 'RAM type', 'RAM frequency', 'HDD', 'SSD',
                            'Video card', 'Model of video card', 'Video card capacity', 'Web-camera', 'Shape', 'Weight'])

def get_info(url):
    response = requests.get(url).text

    soup = BeautifulSoup(response, 'lxml')
    table = soup.find('ul', class_='points-list').find_all('li')
    data = []
    for tab in table:
        data.append({
            'title':tab.find('p', class_='point__left-text').text.strip(),
            'value':tab.find('p', class_='point__right-text').text.strip()
        })
    
    df.loc[url,'Title'] = soup.find('div', class_='mark').find('h1').text.strip()
    try:
        df.loc[url,'Old price'] = soup.find('div', class_='col-left').find('div', class_='price__not-relevant').find_all('span')[0].text.strip()
    except:
        df.loc[url,'Old price'] = ''
    df.loc[url,'New price'] = soup.find('div', class_='col-left').find('div', class_='price__relevant').find_all('span')[0].text.strip()
    df.loc[url,'Image'] = soup.find('div', class_='img-wrapper').find('img').get('src')
    
    for i in range(len(data)):
        if 'Диагональ дисплея' in data[i]['title']: df.loc[url,'Diagonal of display'] = data[i]['value']
        elif 'Разрешение дисплея' in data[i]['title']: df.loc[url,'Shape of display'] = data[i]['value']
        elif 'Серия процессора' in data[i]['title']: df.loc[url,'Processor series'] = data[i]['value']
        elif 'Модель процессора' in data[i]['title']: df.loc[url,'CPU'] = data[i]['value']
        elif 'Количество ядер' in data[i]['title']: df.loc[url,'Cores'] = data[i]['value']
        elif 'Базовая частота процессора' in data[i]['title']: df.loc[url,'CPU base frequency'] = data[i]['value']
        elif 'Объем ОЗУ' in data[i]['title']: df.loc[url,'RAM'] = data[i]['value']
        elif 'Тип памяти' in data[i]['title']: df.loc[url,'RAM type'] = data[i]['value']
        elif 'Частота памяти' in data[i]['title']: df.loc[url,'RAM frequency'] = data[i]['value']
        elif 'Объем жесткого диска HDD' in data[i]['title']: df.loc[url,'HDD'] = data[i]['value']
        elif 'Объем SSD диска' in data[i]['title']: df.loc[url,'SSD'] = data[i]['value']
        elif 'Производитель видеокарты' in data[i]['title']: df.loc[url,'Video card'] = data[i]['value']
        elif 'Модель видеокарты' in data[i]['title']: df.loc[url,'Model of video card'] = data[i]['value']
        elif 'Объем видеопамяти' in data[i]['title']: df.loc[url,'Video card capacity'] = data[i]['value']
        elif 'Веб-камера' in data[i]['title']: df.loc[url,'Web-camera'] = data[i]['value']
        elif 'Габариты' in data[i]['title']: df.loc[url,'Shape'] = data[i]['value']
        elif 'Вес' in data[i]['title']: df.loc[url,'Weight'] = data[i]['value']
        

def main():
    for url in urls:
        get_info(url)

        
if __name__=='__main__':
    main()
```
